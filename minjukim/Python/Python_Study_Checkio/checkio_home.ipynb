# time converter 24 to 12

```python
def time_converter(time):
    hour = time[:2]
    answer = ''
    
    if int(hour) > 12:
        answer += str(int(hour) - 12)
        answer += time[2:] + ' p.m.'
    
    elif int(hour) == 12:
        answer += time + ' p.m.'
    
    elif int(hour) == 0:
        answer += '12'
        answer += time[2:] + ' a.m.'
    
    else: answer += time + ' a.m.'
    
    if answer[0] == '0': return answer[1:]
    return answer
```

```python
time_converter('12:30') 
# time_converter('09:00')
time_converter('23:15')
```

```python
assert time_converter('12:30') == '12:30 p.m.'
assert time_converter('09:00') == '9:00 a.m.'
assert time_converter('23:15') == '11:15 p.m.'
```

# The Most Wanted Letter

```python
def checkio(text):
    text_clean = ''.join([x.lower() for x in text if x.isalpha()])
    
    count_letters = {}
    for letter in text_clean:
        count_letters[letter] = text_clean.count(letter)

    countings = list(count_letters.values())
    if countings.count(max(countings)) > 1:
        
        answer_index = []
        for i, x in enumerate(countings):
            if x == max(countings):
                answer_index.append(i)
        return sorted([list(count_letters.keys())[i] for i in answer_index])[0]
    
    else: return max(count_letters, key=count_letters.get)
```

```python
checkio("Hello World!")
checkio("One")
checkio("AAaooo!!!!")
```

```python
assert checkio("Hello World!") == "l", "Hello test"
assert checkio("How do you do?") == "o", "O is most wanted"
assert checkio("One") == "e", "All letter only once."
assert checkio("Oops!") == "o", "Don't forget about lower case."
assert checkio("AAaooo!!!!") == "a", "Only letters."
assert checkio("abe") == "a", "The First."
```

# Sort Array by Element Frequency

```python
# 푸는 중

def frequency_sort(array):
    array.sort()

    count_elements = {}
    for element in array:
        count_elements[element] = array.count(element)
    
    return count_elements.values()
```

```python
frequency_sort([4, 6, 2, 2, 6, 4, 4, 4])
```

```python
assert list(frequency_sort([4, 6, 2, 2, 6, 4, 4, 4])) == [4, 4, 4, 4, 6, 6, 2, 2]
assert list(frequency_sort(['bob', 'bob', 'carl', 'alex', 'bob'])) == ['bob', 'bob', 'bob', 'carl', 'alex']
assert list(frequency_sort([17, 99, 42])) == [17, 99, 42]
assert list(frequency_sort([])) == []
assert list(frequency_sort([1])) == [1]
```

# Sun Angle

```python
def sun_angle(time):
    hour = int(time[:2])
    minute = int(time[3:])
    
    time_encode = hour + minute / 60
    if (time_encode < 6) | (time_encode > 18): return "I don't see the sun!"
    
    return (time_encode - 6) / 12 * 180
```

```python
assert sun_angle("07:00") == 15
assert sun_angle("01:23") == "I don't see the sun!"
```

# Long Repeat

```python
def long_repeat(line: str) -> int:
    if line:
        answer = []
        count = 0
        former = line[0]
        for i, x in enumerate(line):
            if former != x:
                count = 1
                former = x
            else: count += 1
            answer.append(count) 
        return max(answer)
    return 0
```

```python
# long_repeat('sdsffffse')
# long_repeat('ddvvrwwwrggg')
long_repeat('abababa')
```

```python
assert long_repeat('sdsffffse') == 4, "First"
assert long_repeat('ddvvrwwwrggg') == 3, "Second"
assert long_repeat('abababaab') == 2, "Third"
assert long_repeat('') == 0, "Empty"
```

# Pawn Brotherhood

```python
def safe_pawns(pawns: set) -> int:
    import string
    alpha = [x for x in string.ascii_lowercase]
# 숫자 -1, 알파벳 앞뒤 중에 하나 있으면 safe
    answer = []
    for pawn in pawns:
        for x in pawns:
            if x != pawn:
                alpha_index = alpha.index(pawn[0])
                if (alpha[alpha_index + 1] == x[0]) | (alpha[alpha_index - 1] == x[0]):
                    if int(pawn[1]) - 1 == int(x[1]):
                        answer.append(pawn)

    return len(set(answer))
```

```python
# safe_pawns({"b4", "d4", "f4", "c3", "e3", "g5", "d2"})
safe_pawns({"b4", "c4", "d4", "e4", "f4", "g4", "e5"})
```

```python
assert safe_pawns({"b4", "d4", "f4", "c3", "e3", "g5", "d2"}) == 6
assert safe_pawns({"b4", "c4", "d4", "e4", "f4", "g4", "e5"}) == 1
```

```python
def frequency_sort(items):
    # 원소, 갯수
    count = []
    for x in set(items):
        count.append((x, items.count(x)))

    count_key = [x[0] for x in count]
    count_value = [x[1] for x in count]

    count_order = sorted(count_value, reverse=True)

    # 같은 갯수가 2 개 이상일 때 어떻게 해야할까
    answer = ''
    for c in count_order:
        print(c, count_key[count_value.index(c)])

        answer += str(count_key[count_value.index(c)]) * c
    return [int(x) for x in answer]
```

# Warrior (푸는 중)

```python
class Warrior:

    def __init__(self):
        self.health = 50
        self.attack = 5
        self.is_alive = self.health > 0
        
#     def is_alive(self):
#         return self.health > 0

class Knight(Warrior):
    def __init__(self):
        super().__init__()
        self.attack = 7

def fight(unit_1, unit_2):
    while (unit_1.is_alive and unit_2.is_alive):
        print(unit_1.is_alive, unit_2.is_alive)
        print('alive')
                               
        unit_2.health -= unit_1.attack
        print('attack 1', unit_1.health, unit_2.health)
#         return unit_1.is_alive
                               
        if unit_2.is_alive:
            unit_1.health -= unit_2.attack
            print('attack 2', unit_1.health, unit_2.health)
            
    return unit_1.is_alive
```

```python
chuck = Warrior()
bruce = Warrior()
carl = Knight()
dave = Warrior()
mark = Warrior()
```

```python
fight(chuck, bruce)
```
