# Parameter 모수: 설정값

```python
개판 = []
for i in range(100):
    한계 = 3
    옐로카드 = 0
    짜증확률 = .5
    기간 = 0
    while 옐로카드 < 한계:
        짜증 = binomial(1, 짜증확률)
        if 짜증 == 1:
            옐로카드 += 1
        else:
            옐로카드 = 0
        기간 += 1
    기간
    개판.append(기간)

sns.distplot(개판)
```

### 독립
- 두 사건 $A, B$ 가 독립이면 $P(A | B) = P(A)$ 이고, $P(B | A) = P(B)$
- $P(A, B) = P(A | B) * P(B) = P(A) * P(B)$

# 그래프 모델
-> 데이터로 입증하기 위해 그래프 모델을 만든다

그래프 모델에서 서로 영향을 주고 받는 건(순환고리) 표시할 수 없다.
- 실제로 나눌 수 있음
> ex. 매출과 투자가 서로 영향을 주고 받는 것 같지만
>- 작년 매출이 올해 투자에 영향을 주고 올해의 투자가 내년의 매출에 영향을 준다.
>- (시점의 차이)

# 조건부독립
### B 가 주어졌을 때

$(A)$ → $(B)$ → $(C)$ : $A$ 는 몰라도 $C$ 알 수 있다

$P(A, C | B) = P(C | B) * P(B | A) * P(A) / P(B) = P(C | B) * P(A | B)$

# 로지스틱 선형 모형

- -oid:-를 닮은

```python
from scipy.special import expit
```

```python
expit(0)
```

```python
expit(10)
```

```python
expit(-6)
```

```python
# 선형모형
연기의영향 = 0.01 * 연기력 - .5

sns.distplot(연기의영향)
```

```python
# 로지스틱 선형모형
# 0 ~ 1 사이로 바뀐다

expit(연기의영향)
sns.distplot(expit(연기의영향))
```

```python
주연상 = binomial(5, expit(연기의영향))
```

```python
sns.distplot(주연상)
```

# 선형모형: 연속형 -> 이산형

```python
실력 = normal(50, 15, 100)
실력의영향 = expit(0.5 * 실력 - 25)
합격 = binomial(1, 실력의영향)
```

```python
sns.distplot(실력)
```

```python
sns.distplot(합격)
```

```python
# 결과가 0 아니면 1 
sns.jointplot(실력, 합격)
```

# 선형모형: 이산형 -> 연속형

```python
# 80 % 의 확률로 작동
# 100 대
공기청정기 = binomial(1, .8, 100)
청정기의영향 = 10 * 공기청정기 + 8
공기질 = normal(청정기의영향, 30)
```

```python
sns.distplot(공기질)
```

# 로지스틱 선형모형: 이산형 -> 이산형
연속변수가 이산변수에 영향을 줄 때 쓸 수 있는 테크닉

```python
if 공기청정기 == 1:
    청정기의영향 = 0.73
else:
    청정기의영향 = 0.26
```

위와 같은 내용을 if 문 없이 로지스틱 선형모형으로 만들 수 있다

```python
공기청정기 = binomial(1, .8, 100)
청정기의영향 = expit(2 * 공기청정기 -1)
알러지 = binomial(1, 청정기의영향)

sns.distplot(알러지)
```
