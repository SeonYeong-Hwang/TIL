# Negative Binomial 음이항분포
- 기하분포의 일반화
- 성공률이 $p$ 일 때 $n$ 번 성공할 때까지 시도했을 때 총 시도 횟수
- cf. Geometric 은 $1$ 번 성공할 때까지의 기간만 출력 가능

```python
# 세번 짜증나면 나감

짜증 = 0
기간 = 0

while 짜증 < 3:             # 세번째는 못 참는다
    사건 = binomial(1, .05) # 5 % 의 확률로 사건 발생
    if 사건 == 1:           # 사건이 나면
        짜증 += 1           # 짜증이 증가
    기간 += 1
기간
```

# 포아송 분포
- 평균적으로 $L$번 일어나는 일이 $X$ 번 일어날 확률의 분포
- 이항분포에서 $n$ 이 크고 $p$ 가 작으면 포아송 분포와 비슷해진다
#### ex. 1시간에 평균 10명의 고객이 방문한다고 할 때 시간당 방문 고객 수

```python
from numpy.random import poisson
```

```python
# 평균 30 명이 방문하는 박물관의 50일간 입장객 수 
poisson(30, 50)
```

```python
# 최대 100 명 중 방문확률
# 여기서 100 명(n) 이 숫자가 커지면 포아송 분포와 비슷해진다
# n 은 각각의 독립변수, 동전 던지기의 경우 던지는 동전의 갯수
binomial(100, .3, 50)
```

```python
# 원점을 둘러싼 한 변이 2인 정사각형(+-1, +-1)
# 원점을 둘러싼 대각선이 2인 마름모(+-1, +-1)
# 정사각형 안의 점을 하나 뽑았을 때 마름모 안에 들어갈 확률(1/2)

x = uniform(-1, 1, 1000)
y = uniform(-1, 1, 1000)

# |x| + |y| < 1
np.mean(np.abs(x) + np.abs(y) < 1)
```

```python
# 원점을 둘러싼 한 변이 2인 정사각형(+-1, +-1)
# 원점을 둘러싼 지름이 2(반지름이 1)인 원
# 정사각형 안의 점을 하나 뽑았을 때 마름모 안에 들어갈 확률(3.14 / 4)
x = uniform(-1, 1, 10000000)
y = uniform(-1, 1, 10000000)

np.mean(x ** 2 + y ** 2 < 1) * 4
```

# 정규분포
- 평균 $m$, 분산 $s²$ 을 모수로 하는 연속 확률 분포
- 이항 분포에서 $n$ 이 커지면 정규 분포와 비슷해진다

- 가장 자연스러운 분포인 것은 아님, 그냥 다루기 쉽다

```python
from numpy.random import normal
```

```python
# 고객 100 명이 평균 50 만원 쓸 때(표준편차 20 만원)
하루매출 = normal(50, 20, 100)
```

```python
# 기하분포
기간 = geometric(.05, 100)
```

```python
기간 * 하루매출
```

```python
# 0 이면 light user, 1 이면 heavy user
# 괄호 안 첫번째 값(1) 던지는 동전의 갯수
# 괄호 안 두번째 값(0.01) 은 heavy user 의 비율
유저집단 = binomial(1, .001, 100000)
```

```python
총매출list = []
총매출 = 0

for 유저 in 유저집단:
    if 유저 == 0: # light user
        매출 = normal(1, 10)
    else: # heavy user
        매출 = normal(100, 10)
    총매출list.append(매출)
    총매출 += 매출
```

```python
print(총매출)
총매출list
```

# 평균으로의 회귀
- 위의 예시에서 한국인이 나올 확률 50% 확률인데, 더 큰 수도 나오고 더 작은 수도 나온다
- 더 큰 수가 나오면 그 다음날은 더 작은 수가 나온다
- 큰 수가 나오면 그 수보다 큰 수가 나오기 어렵기 때문에 그 수보다 작은 수가 나온다
- 결과적으로 오르락 내리락하는 패턴이 있는 것처럼 보인다

# 지수분포 Exponential
- 다음번 일어날 때까지 얼마나 걸리는지
- ex. 다음 버스가 올 때까지, 다음 지진이 일어날 때까지 얼마나 걸리는지
- 
- cf. poission 포아송 분포: 그 일이 일어날 때까지 얼마나 걸리는지

```python
from numpy.random import exponential
```

```python
exponential(30)
```

```python
# 리드 타임: 고객이 도착할 때부터 서비스까지 걸리는 시간
리드타임 = 10
# 평균 30분 후에 고객 도착
# 100 명의 고객
np.mean(exponential(30, 100) < 리드타임) # 리드타임 내 추가 고객 도착 비율

# 10명까지 대기할 수 있는 경우
# 리드타임을 얼마로 해야 최적화되는가
```

```python
# 리드타임은 같은데, 알바가 여러 명인 경우
알바 = 10
리드타임 = 10 / 알바
np.mean(exponential(30, 100) < 리드타임) # 리드타임 내 추가 고객 도착 비율
```
